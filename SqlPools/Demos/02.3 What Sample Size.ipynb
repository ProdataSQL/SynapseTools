## 2.3 What Samples Size should I used ?

The default sample size is 20%, **<mark>BUT MS recommend 2% when \> 1 billion row</mark>**. So this means you may want to avoid defaults on large tables.

An adaptive algothim commonly used is below

<span lang="EN-GB" style="font-size:17.5pt;font-family:&quot;Segoe UI&quot;,sans-serif;
color:#0F1419;background:white">SQRT([Row Count])*1000/[Row Count]</span>
